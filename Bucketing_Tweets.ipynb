```python
#Import Libraries
import os
import pandas as pd
from pandas.core.indexes.base import Index

os.chdir(r'C:\Users\mbhar\Downloads\twitter_project')


#reading tweets file on which sentimental analysis was performed
sentiments = pd.read_csv(r'sentiment.csv')

#converting the tweets datatype to string as it currently is in Object because of processing while deriving sentiment in sentiment analysis code
sentiments['tweets'] = sentiments['tweets'].astype('str')


#Funtion to extract hashtags from tweets
def extract_hashtags(text):
     
    # initializing hashtag_list variable
    hashtag_list = []
     
    # splitting the text into words
    for word in text.split():
         
        # checking the first character of every word
        if word[0] == '#':
             
            # adding the word to the hashtag_list
            hashtag_list.append(word[1:])
    return hashtag_list

#extracting hashtags using above function and saving in a new column
sentiments['hastags'] = sentiments['tweets'].apply(extract_hashtags)



#bitcoin

#creating flag column to be used for filtering in next steps
sentiments['bit']=''

#creating array list for bitcoin hashtags
haslist=['bitcoin','Bitcoin','BTC','btc','bitcoins','bitcoinexchange','bitcoincash']

#checking if hashtags column contains any of the terms from above array list
for index, row in sentiments.iterrows():
    res = [ele for ele in haslist if(ele in row['hastags'])]
 
    if len(res) > 0:
           sentiments['bit'][index]='yes'
    else :
        sentiments['bit'][index]='no'

#filtering tweets based on hashtag
filtered_hastags=sentiments[sentiments['bit']=='yes']


#filtering tweets based on tweet text using regex
filtered_bitcoin=sentiments[sentiments['tweets'].str.contains('bitcoin|btc|bitcoins|bitcoinexchange|bitcoincash')]

#removing duplicate rows based on username and time
sentiment_bitcoin = filtered_bitcoin.drop_duplicates(subset=['name','time'],ignore_index= True)

#saving the tweets as CSV
sentiment_bitcoin.drop(columns=['bit']).to_csv("bitcoin_sentiment.csv", header= True, index = False)


#ethereum

sentiments['eth']=''

haslist=['ethereum','Ethereum','eth','ETH','ethereumprice','ethbtc']


for index, row in sentiments.iterrows():
    res = [ele for ele in haslist if(ele in row['hastags'])]
 
    if len(res) > 0:
           sentiments['eth'][index]='yes'
    else :
        sentiments['eth'][index]='no'

filtered_hastags=sentiments[sentiments['eth']=='yes']


filtered_ethereum=sentiments[sentiments['tweets'].str.contains('ethereum|Ethereum|eth|ETH|ethereumprice|ethbtc')]

sentiment_ethereum = filtered_ethereum.drop_duplicates(subset=['name','time'],ignore_index= True)

sentiment_ethereum.drop(columns=['eth']).to_csv("ethereum_sentiment.csv", header= True, index = False)

#binance

sentiments['bnb']=''

haslist=['binance','binancecoin','BinanceCoin','bnb','BNB','Binance']


for index, row in sentiments.iterrows():
    res = [ele for ele in haslist if(ele in row['hastags'])]
 
    if len(res) > 0:
           sentiments['bnb'][index]='yes'
    else :
        sentiments['bnb'][index]='no'

filtered_hastags=sentiments[sentiments['bnb']=='yes']


filtered_binance=sentiments[sentiments['tweets'].str.contains('binance|binancecoin|BinanceCoin|bnb|BNB|Binance')]

sentiment_binance = filtered_binance.drop_duplicates(subset=['name','time'],ignore_index= True)

sentiment_binance.drop(columns=['bnb']).to_csv("binance_sentiment.csv", header= True, index = False)
```
